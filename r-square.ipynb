<h1>An Other Example</h1>


R&sup2; is very similar to its cousin, R, but it is more easy to Interpretation. For eaxample it is not obvious that R = 0.7 is twice as good as R = 0.5 


In Summary, The value for R&sup2; can range from 0 to 1. A value of 0 indicates that the response variable cannot be explained by the predictor variable at all while a value of 1 indicates that the response variable can be perfectly explained by the predictor variables.

There is a problem with the R&sup2;. 
<span style="font-size: 40px;">&#x1F629;</span>


**R&sup2;** only works as intended in a simple linear regression model with one explanatory variable. With a multiple regression made up of several independent variables, the R&sup2; must be adjusted.

**R&sup2;** increases or remains the same as we add new predictors to the multiple regression model. This remains the case even if the newly added predictors are independent of the target variable and dont add any value to the predicting power of the model. Adjusted **R&sup2;** only increases if the newly added predictor improves the models predicting power. 

In the other word, How many junk independent variables or important independent variable or impactful independent variable you add to your model, the **R&sup2;** value will 
always increase. It will never decrease with the addition of a newly independent variable, whether it could 
be an impactful, non-impactful, or bad variable, so we need another way to measure equivalent **R&sup2;**, which penalizes our model with any junk independent variable.

In the example below, in model 2, when we add Temperature (as impactful independent varibles which is is useless in predicting the price of a pizza) it increased the **R&sup2;** but the **adjusted R&sup2;** decreased. 



```python
import numpy as np
import pandas as pd
import statsmodels.api as sm

# Sample data for price of dough, price of pizza, and temperature
data = {
    'Price_Dough': [10, 20, 30, 40, 50],
    'Price_Pizza': [100, 180, 250, 310, 380],
    'Temperature': [25, 30, 28, 35, 32],
}

df = pd.DataFrame(data)

# Separate independent variable (X1) and dependent variable (Y)
X1 = df['Price_Dough']
Y = df['Price_Pizza']

# Add constant for intercept term in the model
X1 = sm.add_constant(X1)

# Fit the linear regression model with X1
model_1 = sm.OLS(Y, X1).fit()

# Get R-squared and Adjusted R-squared values for model with X1
R_squared_1 = model_1.rsquared
Adjusted_R_squared_1 = model_1.rsquared_adj

print("Model 1 - Price of Dough:")
print("R-squared:", R_squared_1)
print("Adjusted R-squared:", Adjusted_R_squared_1)

# Now, add temperature (X3) as an additional independent variable
X3 = df['Temperature']
X2 = pd.concat([X1, X3], axis=1)  # Combine X1 and X3

# Fit the linear regression model with X1 and X3
model_2 = sm.OLS(Y, X2).fit()

# Get R-squared and Adjusted R-squared values for model with X1 and X3
R_squared_2 = model_2.rsquared
Adjusted_R_squared_2 = model_2.rsquared_adj

print("\nModel 2 - Price of Dough and Temperature:")
print("R-squared:", R_squared_2)
print("Adjusted R-squared:", Adjusted_R_squared_2)
```
